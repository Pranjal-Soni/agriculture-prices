```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import warnings
import sys
import warnings
if not sys.warnoptions:
    warnings.simplefilter("ignore")
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df = pd.read_excel("patato_prices.xlsx")
```

```python
df
```

```python
market_names = df["Market Name"].unique()
market_names
```

```python
for mname in market_names:
    sub_df = df[df["Market Name"]==mname]
    plt.figure(figsize=(8,6))
    sub_df["Price Date"] = pd.to_datetime(sub_df["Price Date"])
    sns.lineplot(x = "Price Date", y = "Min Price (Rs./Quintal)", data = sub_df,)
    sns.lineplot(x = "Price Date", y = "Max Price (Rs./Quintal)", data = sub_df)
    sns.lineplot(x = "Price Date", y = "Modal Price (Rs./Quintal)", data = sub_df)
    plt.ylabel("Price (Rs./Quintal)")
    plt.xticks(rotation =25)
    plt.legend(["Min Price (Rs./Quintal)","Max Price (Rs./Quintal)","Modal Price (Rs./Quintal)"], loc='center left', bbox_to_anchor=(1, .5))
    plt.title(mname+" Market Patato Prices")
    plt.show()
```
